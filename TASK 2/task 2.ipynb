```python
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train.head()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
train["Transported"].value_counts()
```

```python
y = train["Transported"]
train = train.drop("Transported", axis=1)
```

```python
num_cols = ["Age","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"]

for col in num_cols:
    train[col].fillna(train[col].median(), inplace=True)
    test[col].fillna(train[col].median(), inplace=True)
```

```python
cat_cols = ["HomePlanet","CryoSleep","Destination","VIP"]

for col in cat_cols:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(train[col].mode()[0], inplace=True)
```

```python
train[['Deck','Num','Side']] = train['Cabin'].str.split("/", expand=True)
test[['Deck','Num','Side']] = test['Cabin'].str.split("/", expand=True)

train.drop("Cabin", axis=1, inplace=True)
test.drop("Cabin", axis=1, inplace=True)
```

```python
train.drop(["Name","PassengerId"], axis=1, inplace=True)
test.drop(["Name","PassengerId"], axis=1, inplace=True)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)

train, test = train.align(test, join="left", axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

preds = model.predict(X_val)

print("Accuracy:", accuracy_score(y_val, preds))
print(classification_report(y_val, preds))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

rf_preds = rf.predict(X_val)
print("RF Accuracy:", accuracy_score(y_val, rf_preds))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "n_estimators": [100,200],
    "max_depth": [None,10,20]
}

grid = GridSearchCV(RandomForestClassifier(random_state=42), params, cv=5)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
```

```python
best_model.fit(train, y)
```

```python
test_preds = best_model.predict(test)
```

```python
submission = pd.read_csv("sample_submission.csv")
submission["Transported"] = test_preds
submission.to_csv("submission.csv", index=False)
```
