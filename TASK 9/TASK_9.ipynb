### **Text Summarization using Hugging Face and Sumy**

```python
text = """Artificial Intelligence (AI) is a transformative branch of computer science dedicated to creating systems capable of performing tasks that
typically require human intelligence, such as reasoning, learning, problem-solving, and perception. By analyzing vast datasets to identify patterns,
AI algorithms can make predictions, automate complex workflows, and enhance decision-making in real-time. Modern AI, which includes subfields like
machine learning (ML) and deep learning, powers daily technologies such as voice assistants, recommendation engines, and autonomous vehicles. While
AI offers significant benefits—including improved efficiency in healthcare diagnostics and increased productivity in manufacturing—it also presents
challenges regarding algorithmic bias, job displacement, and ethical concerns. As it continues to evolve, AI is shifting from narrow, task-specific
applications toward more generative and agentic capabilities, reshaping how industries operate and how humans interact with technology."""
```

**Hugging Face Summarization**

```python
from transformers import pipeline

summarizer = pipeline("summarization")

hf_summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print("Hugging Face Summary:")
print(hf_summary[0]['summary_text'])
```

**Sumy Summarization**

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

parser = PlaintextParser.from_string(text, Tokenizer("english"))
lsa_summarizer = LsaSummarizer()

summary = lsa_summarizer(parser.document, 2)

print("\nSumy Summary:")
for sentence in summary:
    print(sentence)
```
